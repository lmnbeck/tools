```python
from selenium import webdriver
import re
import pandas as pd
import time


# setting chrome
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument("window-size=1024,768")
chrome_options.add_argument("--no-sandbox")
```

```python
# get url
browser = webdriver.Chrome(options = chrome_options)

all_data = pd.DataFrame()
def pa_hanjie(page):
    global all_data 
    url = 'http://www.china-weldnet.com/cgi-bin/qyml/result.asp?page=' + str(page) + '&keyword=%C9%E8%B1%B8'
    browser.get(url)
    time.sleep(2)
    data = browser.page_source
    table = pd.read_html(data)[1]
#     print(table[12:])
    if page>1:
        all_data= all_data.append(table[1:])
    else:
        all_data= all_data.append(table)
    




```

```python

r_pages = '<font color="red">\d*?</font>/(\d*?) 页 &nbsp;总共 '
total_page = re.findall(r_pages, data)
```

```python
for i in range(1,15):
    pa_hanjie(i)

# print(all_data) 
all_data.to_excel("焊接协会会员信息.xlsx", index=False)
```
